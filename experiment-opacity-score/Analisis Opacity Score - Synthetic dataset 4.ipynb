```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

# Logistic Regression

```python
df_lr = pd.read_csv('df_results_lr.csv')
df_lr
```

```python
import numpy as np
import matplotlib.pyplot as plt


sns.set(rc={'figure.figsize':(11,7)})
num_variables = list(range(2,19))
opacity_score = df_lr.mean_trace.values
plt.plot(num_variables, opacity_score, 'o-')
plt.title('Logistic Regression - Opacity Score (Variance of Explanations)')
plt.ylabel('Trace of covariance matrix')
#plt.ylabel('Explanation fit explanation model')
plt.xlabel('Num. variables')
```

```python
import numpy as np
import matplotlib.pyplot as plt


sns.set(rc={'figure.figsize':(11,7)})
num_variables = list(range(2,19))
opacity_score = df_lr.mean_explanation_fit.values
plt.plot(num_variables, opacity_score, 'o-')
plt.title('Logistic Regression - Opacity Score (Explanation Fit)')
#plt.title('Logistic Regression - Opacity Score (Variance of Explanations)')
#plt.ylabel('Trace of covariance matrix')
plt.ylabel('Explanation fit explanation model')

plt.xlabel('Num. variables')
```

```python
import numpy as np
import matplotlib.pyplot as plt


sns.set(rc={'figure.figsize':(11,7)})
num_variables = list(range(2,19))
opacity_score = df_lr.auc_train.values
opacity_score1 = df_lr.auc_test.values
plt.plot(num_variables, opacity_score, 'o-')
plt.plot(num_variables, opacity_score1, 'o-')
plt.title('Logistic Regression - Opacity Score')
plt.ylabel('Max eigenvalue covariance explanations matrix')
plt.xlabel('Num. variables')
```

```python
import numpy as np
import matplotlib.pyplot as plt


sns.set(rc={'figure.figsize':(11,7)})
num_variables = list(range(2,19))
opacity_score = (df_lr.auc_train.values - df_lr.auc_test.values)
opacity_score1 = df_lr.mean_trace.values
plt.plot(num_variables, opacity_score, 'o-')
plt.plot(num_variables, opacity_score1, 'o-')
plt.title('Logistic Regression - Opacity Score')
plt.ylabel('Max eigenvalue covariance explanations matrix')
plt.xlabel('Num. variables')
plt.legend(['overfitting', 'opacity_score'], loc='upper left')
```
