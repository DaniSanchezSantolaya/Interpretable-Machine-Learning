```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy

import matplotlib.pyplot as plt
%matplotlib inline
```

# Load explanations

```python
data = {}

# Explanations logistic regression
data['LR'] = {}
num_vars = list(range(2,19,1))
for n_var in num_vars:
    data['LR'][n_var] = np.genfromtxt('explanations_data/explanations_lr_' + str(n_var) + '_0.csv', delimiter=',')
    
# Explanation decision tree
data['DT'] = {}
l_max_depth = list(range(2, 19, 2))
for max_depth in l_max_depth:
    data['DT'][max_depth] = np.genfromtxt('explanations_data/explanations_dt_' + str(max_depth) + '_0.csv', delimiter=',')

# Explanation random forest
data['RF'] = {}
l_n_trees = [2, 5, 12, 20, 30, 50, 100]
for n_trees in l_n_trees:
    data['RF'][n_trees] = np.genfromtxt('explanations_data/explanations_rf_' + str(n_trees) + '_0.csv', delimiter=',')
    
# Explanation Neural Networks
data['NN'] = {}
l_architectures_1 = [(10,), (50,), (100,)]
for architecture in l_architectures_1:
    data['NN'][architecture] = np.genfromtxt('explanations_data/explanations_nn_' + str(architecture) + '_0.csv', delimiter=',')

l_architectures_2 = [(50,), (50,50,50), (50,50,50,50,50)]
for architecture in l_architectures_2:
    data['NN'][architecture] = np.genfromtxt('explanations_data/explanations_nn5_' + str(architecture) + '_0.csv', delimiter=',')
```

# Show distributions

```python
# Logistic Regression
fig, axarr = plt.subplots(6, 7, figsize=(16, 10))

row = 0
for num_vars in [2,3,4,5,6,7]:
    for i in range(num_vars):
        #sns.distplot(data['LR'][num_vars][:,i], ax=axarr[row][i], bins=10)
        sns.kdeplot(data['LR'][num_vars][:,i], ax=axarr[row][i], shade=True)
        #axarr[row][i].set_title("variable " , fontsize=12)
    row += 1

plt.subplots_adjust(hspace=.3)

import seaborn as sns
sns.despine()
```
